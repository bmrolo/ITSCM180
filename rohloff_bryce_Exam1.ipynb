## Problem 1

For any two points like P1 and P2 with x-y coordinates of P1(x1, y1) and P2(x2, y2), the following formulas can be used to calculate the **Manhattan** and **Euclidean** distance between them. 

Define two python functions called **MTN** and **EUC** that each receive *x1*, *y1*, *x2*, and *y2* as parameters (in the same order) and return the corresponding calculated distance as output.
<br>**Note**: You can use the **abs** function from the python standard library(i.e., no need to import any modules) to calculate the absolute values in the Manhattan distance formula. I assume you already know the appropriate function (and its module) for calculating the square root. You may import the required module in the body of a function definition statement.
<br> To make sure your defined functions work fine, you can call them in a separate coding cell and try the following examples and see if they return the expected output:
<br> MTN(4,0,0,4)  output: 8
<br> EUC(4,0,0,4)  output: 5.65685

```python
import math 
def EUC(x1,x2,y1,y2):
    EDistance = math.sqrt( ((x1-x2)**2)+ ((y1-y2)**2) )
    return EDistance

def MTN(x1,x2,y1,y2):
    MDistance = abs(x1-x2) + abs(y1-y2)
    return MDistance
```

## Problem 2

Write a program that prompts the user to enter the x and y coordinates of two points as well as the desired distance measure (**Manhattan** or **Euclidean**), then depending on the requested measure and using appropriate function from those you defined in Problem 1 calculate the requested distance and show it to the user.

<br>Example:
<br>Enter x coordinate of point1:4
<br>Enter y coordinate of point1:1
<br>Enter x coordinate of point2:3
<br>Enter y coordinate of point2:6
<br>Enter your desired distance measure: Euclidean
<br>The Euclidean distance between the points is: 5.099

```python
# getting info from the user 
x1 = float(input('Enter x coordinate of point1:'))
y1 = float(input('Enter y coordinate of point1:'))
x2 = float(input('Enter x coordinate of point2:'))
y2 = float(input('Enter y coordinate of point2:'))
desired_distance_measure = input('Enter your desired distance measure:')

#Euclidean path
if desired_distance_measure == 'Euclidean':
    distance = EUC(x1,x2,y1,y2)
#Manhattan path
elif desired_distance_measure == 'Manhattan':
    distance = MTN(x1,x2,y1,y2)
print('The', desired_distance_measure , 'distance between the points is:', distance)
```

## Problem 3

Write a program that prompts a student to enter his/her number of credit hours enrolled, annual household income, and the state of residency, then calculates his/her tuition using the rates given in the following table.

In case the student's annual household income was less than 70,000 dollars, apply a %15 discount to the total tuition calculated. 

<br> Example:
<br> Enter the number of credit hours enrolled: 15
<br> What is your state of residency? IL
<br> Enter your annual household income: 66000
<br> Your tuition for this semester is: 7012.5 USD

```python
# getting user input
credits = float(input('Enter the number of credit hours enrolled: '))
SOR = input('What is your state of residency? ')
income = float(input('Enter your annual household income: '))

# setting tuition variable 
tuition = 0 

# using the user input to apply the correct credit rate
if SOR == 'MN' or SOR == 'IL' or SOR == 'MI':
    tuition = credits * 550
elif SOR == 'WI':
    tuition = credits * 425
else:
    tuition = credits * 650

# applying income discount
if income < 70000:
    tuition = tuition * .85

print('Your tuition for this semester is:', tuition, 'USD')
```
